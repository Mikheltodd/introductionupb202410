````python
# Continuación - Operaciones con números

# Suma
a = 3
b = -4
c = a + b
print(a)

# Resta
d = 5
e = 7
f = d - e

# Multiplicación
a = b * c # reasignación de variables
print(a)

# División
a = b / a
print(a)

# Módulo
a = 6
a = a % 2
print(a)

# Potencia
a = 8
print(a ** 2)
print(a ** 3)
print(a ** (1/2))
print(a ** (1/3))
````

````python
# Concatenar

firstname = "Miguel"
lastname = "Ortiz"
fullname = lastname + ", " + firstname
print(fullname)
````

````python
# +=

a = 5
a += 10
print(a)
a -= 8
a *= 3
a /= 2
a **= 2
print(a)
````
